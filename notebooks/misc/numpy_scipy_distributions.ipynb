## Numpy vs Scipy Distributions

A useful reference to help reduce confusion about distribution parameterization. 

```python
import numpy as np
import scipy.stats as st

import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# set random state for repeatability

random_state = np.random.RandomState(42)
n_samples = 100000
```

## Discrete Distributions

```python
# Negative Binomial

n = 100
p = .3

numpy_sample = random_state.negative_binomial(n=n, p=p, size=(n_samples, 1))
dist = st.nbinom(n=n, p=p)
dist.random_state = random_state
scipy_sample = dist.rvs(size=(n_samples, 1))

plt.figure(figsize=(10, 4))
plt.title("Negative Binomial")
sns.distplot(numpy_sample)
sns.distplot(scipy_sample)
plt.show()

statistic, pvalue = st.ks_2samp(numpy_sample[:, 0], scipy_sample[:, 0])
assert pvalue >= 0.05, print(pvalue)

# Poisson

lam = 2.0 # rate

numpy_sample = random_state.poisson(lam=lam, size=(n_samples, 1))
dist = st.poisson(mu=lam)
dist.random_state = random_state
scipy_sample = dist.rvs(size=(n_samples, 1))

plt.figure(figsize=(10, 4))
plt.title("Poisson")
sns.distplot(numpy_sample)
sns.distplot(scipy_sample)
plt.show()

statistic, pvalue = st.ks_2samp(numpy_sample[:, 0], scipy_sample[:, 0])
assert pvalue >= 0.05, print(pvalue)

# Binomial

n = 100
p = .3

numpy_sample = random_state.binomial(n=n, p=p, size=(n_samples, 1))
dist = st.binom(n=n, p=p)
dist.random_state = random_state
scipy_sample = dist.rvs(size=(n_samples, 1))

plt.figure(figsize=(10, 4))
plt.title("Binomial")
sns.distplot(numpy_sample)
sns.distplot(scipy_sample)
plt.show()

statistic, pvalue = st.ks_2samp(numpy_sample[:, 0], scipy_sample[:, 0])
assert pvalue >= 0.05, print(pvalue)

# Hypergeometric

N = 5000
M = ngood = 1000
nbad = N - M
n = 2000

numpy_sample = random_state.hypergeometric(ngood=ngood, nbad=nbad, nsample=n, size=n_samples)
dist = st.hypergeom(M=M, n=n, N=N)
dist.random_state = random_state
#scipy_sample = dist.rvs(size=n_samples)

plt.figure(figsize=(10, 4))
plt.title("Hypergeometric")
sns.distplot(numpy_sample)
#sns.distplot(scipy_sample)
plt.show()

#statistic, pvalue = st.ks_2samp(numpy_sample, scipy_sample)
#assert pvalue >= 0.05, print(pvalue)

# Multinomial

# Dirichlet
```

## Continuous Distributions

```python
# Gamma

k = alpha = 3.0 # shape
beta = 5 # rate (inverse scale)
theta = 1.0 / beta # scale

numpy_sample = random_state.gamma(shape=alpha ,scale=theta, size=(n_samples, 1))
dist = st.gamma(a=alpha, scale=beta)
dist.random_state = random_state
scipy_sample = dist.rvs(size=(n_samples, 1))

plt.figure(figsize=(10, 4))
plt.title("Gamma")
sns.distplot(numpy_sample)
sns.distplot(scipy_sample)
plt.show()

statistic, pvalue = st.ks_2samp(numpy_sample[:, 0], scipy_sample[:, 0])
assert pvalue >= 0.05, print(pvalue)

# Inverse Gamma

alpha = 3.0 # shape
beta = 5.0 # scale

dist = st.invgamma(a=alpha, scale=beta)
dist.random_state = random_state
scipy_sample = dist.rvs(size=(n_samples, 1))

plt.figure(figsize=(10, 4))
plt.title("Inverse Gamma")
sns.distplot(scipy_sample)
plt.show()

# Exponential

lam = alpha = 3.0 # scale

numpy_sample = random_state.exponential(scale=lam, size=(n_samples, 1))
dist = st.expon(scale=lam)
dist.random_state = random_state
scipy_sample = dist.rvs(size=(n_samples, 1))

plt.figure(figsize=(10, 4))
plt.title("Exponential")
sns.distplot(numpy_sample)
sns.distplot(scipy_sample)
plt.show()

statistic, pvalue = st.ks_2samp(numpy_sample[:, 0], scipy_sample[:, 0])
assert pvalue >= 0.05, print(pvalue)

# Weibull

k = beta = 5.0 # shape
lam = theta = 3.0 # scale

numpy_sample = np.random.weibull(a=k, size=(n_samples, 1)) * lam
dist = st.weibull_min(c=k, loc=0, scale=lam)
dist.random_state = random_state
scipy_sample = dist.rvs(size=(n_samples, 1))

plt.figure(figsize=(10, 4))
plt.title("Weibull")
sns.distplot(numpy_sample)
sns.distplot(scipy_sample)
plt.show()

statistic, pvalue = st.ks_2samp(numpy_sample[:, 0], scipy_sample[:, 0])
assert pvalue >= 0.05, print(pvalue)

# Pareto

alpha = 3.0 # shape
x_m = 1.0 # scale

numpy_sample = np.random.pareto(a=alpha, size=(n_samples, 1)) + x_m
dist = st.pareto(b=alpha, scale=x_m)
dist.random_state = random_state
scipy_sample = dist.rvs(size=(n_samples, 1))

plt.figure(figsize=(10, 4))
plt.title("Pareto")
sns.distplot(numpy_sample)
sns.distplot(scipy_sample)
plt.show()

statistic, pvalue = st.ks_2samp(numpy_sample[:, 0], scipy_sample[:, 0])
assert pvalue >= 0.05, print(pvalue)

# Normal

mu = 10 # mean
var = sigma_squared = inverse_tau = 10# variance / inverse precision

numpy_sample = np.random.normal(loc=mu, scale=np.sqrt(var), size=(n_samples, 1))
dist = st.norm(loc=mu, scale=np.sqrt(var))
dist.random_state = random_state
scipy_sample = dist.rvs(size=(n_samples, 1))

plt.figure(figsize=(10, 4))
plt.title("Normal")
sns.distplot(numpy_sample)
sns.distplot(scipy_sample)
plt.show()

statistic, pvalue = st.ks_2samp(numpy_sample[:, 0], scipy_sample[:, 0])
assert pvalue >= 0.05, print(pvalue)

# Student T

mu = 10 # mean
var = sigma_squared = inverse_tau = 10# variance / inverse precision
nu = 3 # degrees of freedom

numpy_sample = np.random.standard_t(df=nu, size=(n_samples, 1)) * np.sqrt(var) + mu
dist = st.t(loc=mu, scale=np.sqrt(var), df=nu)
dist.random_state = random_state
scipy_sample = dist.rvs(size=(n_samples, 1))

plt.figure(figsize=(10, 4))
plt.title("Student T")
sns.distplot(numpy_sample)
sns.distplot(scipy_sample)
plt.show()

statistic, pvalue = st.ks_2samp(numpy_sample[:, 0], scipy_sample[:, 0])
assert pvalue >= 0.05, print(pvalue)
```
