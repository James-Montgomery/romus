# Continuous Likelihood Model Examples

Here we demonstrate the use of the various continuous models implemented in this package. We include basic heuristic examples of data that could be analyzed using these models. 

```python
import numpy as np
#import scipy.stats as st

import matplotlib.pyplot as plt
import seaborn as sns

import romus

np.random.seed(42)
```

## Gamma-Exponential Model

...

```python
data_size = 500
data = np.random.exponential(scale=5, size=(data_size, 1))

model = romus.continuous_likelihood.GammaExponential(1.0, 1.0)
prior_samples = model.sample_prior(1000)
prior_predictive_sample_means = [
    model.sample_prior_predictive(data_size).mean()
    for _ in range(1000)
]
prior_predictive_sample_stds = [
    model.sample_prior_predictive(data_size).std()
    for _ in range(1000)
]
model.fit(data=data)
posterior_samples = model.sample_posterior(1000)
posterior_predictive_sample_means = [
    model.sample_posterior_predictive(data_size).mean()
    for _ in range(1000)
]
posterior_predictive_sample_stds = [
    model.sample_posterior_predictive(data_size).std()
    for _ in range(1000)
]

plt.figure(figsize=(15,4))

plt.subplot(211)
sns.distplot(prior_samples, label="Prior")
sns.distplot(posterior_samples, label="Posterior")
plt.axvline(1/5, c="k", label="True Value")
plt.xlim(0, 1)
plt.legend()

plt.subplot(223)
sns.distplot(prior_predictive_sample_means, label="Prior Predictive Means")
sns.distplot(posterior_predictive_sample_means, label="Posterior Predictive Means")
plt.axvline(data.mean(), c="k", label="Data")
plt.xlim(-10, 10)
plt.legend()

plt.subplot(224)
sns.distplot(prior_predictive_sample_stds, label="Prior Predictive Stds")
sns.distplot(posterior_predictive_sample_stds, label="Posterior Predictive Stds")
plt.axvline(data.std(), c="k", label="Data")
plt.xlim(0, 10)
plt.legend()

plt.show()
```

## Normal-Gamma-Normal Model

...

```python
data_size = 2000
data = np.random.normal(loc=5, scale=np.sqrt(1.0/.1), size=(data_size, 1))

model = romus.continuous_likelihood.NormalGammaNormal(1.0, 1.0, 0, 1)
mu_prior_samples, tau_prior_samples = model.sample_prior(1000)
prior_predictive_sample_means = [
    model.sample_prior_predictive(data_size).mean()
    for _ in range(1000)
]
prior_predictive_sample_stds = [
    model.sample_prior_predictive(data_size).std()
    for _ in range(1000)
]
alpha_s, beta_s, mu_s, nu_s = model.fit(data=data)
mu_posterior_samples, tau_posterior_samples = model.sample_posterior(1000)
posterior_predictive_sample_means = [
    model.sample_posterior_predictive(data_size).mean()
    for _ in range(1000)
]
posterior_predictive_sample_stds = [
    model.sample_posterior_predictive(data_size).std()
    for _ in range(1000)
]

plt.figure(figsize=(15,4))

plt.subplot(221)
sns.distplot(mu_prior_samples, label="Prior Mu")
sns.distplot(mu_posterior_samples, label="Posterior Mu")
plt.axvline(5, c="k", label="True Value")
plt.xlim(-10, 10)
plt.legend()

plt.subplot(222)
sns.distplot(tau_prior_samples, label="Prior Tau")
sns.distplot(tau_posterior_samples, label="Posterior Tau")
plt.axvline(.1, c="k", label="True Value")
plt.xlim(0, 0.5)
plt.legend()

plt.subplot(223)
sns.distplot(prior_predictive_sample_means, label="Prior Predictive Means")
sns.distplot(posterior_predictive_sample_means, label="Posterior Predictive Means")
plt.axvline(data.mean(), c="k", label="Data")
plt.xlim(-10, 10)
plt.legend()

plt.subplot(224)
sns.distplot(prior_predictive_sample_stds, label="Prior Predictive Stds")
sns.distplot(posterior_predictive_sample_stds, label="Posterior Predictive Stds")
plt.axvline(data.std(), c="k", label="Data")
plt.xlim(0, 10)
plt.legend()

plt.show()
```
